## Payment Table Relationships

The `payment` table contains transaction-style data.

### Primary Key

- `payment_id`

### Foreign Keys

- `customer_id` → `customer.customer_id`
- `rental_id` → `rental.rental_id`
- `staff_id` → `staff.staff_id`

### Financial Field

- `amount DECIMAL(5,2)`

This is important because financial values should usually use `DECIMAL`, not floating-point types.

### Relationship Diagram

payment
├── payment_id (Primary Key)
├── customer_id → customer
├── rental_id → rental
├── staff_id → staff
├── amount
└── payment_date

## Market Analysis Analogy

The `payment` table is similar to a transaction table in a market database.

| Sakila Payment Table | Market Database Example |
|---|---|
| payment_id | trade_id |
| customer_id | account_id |
| amount | trade_value |
| payment_date | trade_date |
| rental_id | transaction_reference_id |

The same design logic could apply to:

- stock trades
- portfolio transactions
- dividend payments
- cash flows
- historical price observations
