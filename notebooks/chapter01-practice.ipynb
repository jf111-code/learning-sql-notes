## Customer Table Relationships

The customer table contains:

### Primary Key

customer_id

Purpose:
Uniquely identifies each customer.

### Foreign Keys

address_id → address.address_id

store_id → store.store_id

Relationship Diagram

customer
├── customer_id (Primary Key)
├── address_id (Foreign Key)
└── store_id (Foreign Key)

customer
    ↓
 address
    ↓
  city
    ↓
 country

customer
    ↓
  store

## Market Analysis Analogy

Imagine building a stock market database.

company
├── company_id (Primary Key)
├── sector_id (Foreign Key)
└── exchange_id (Foreign Key)

company
    ↓
 sector

company
    ↓
 exchange

The same relational principles used in Sakila can be used for:

- company databases
- stock prices
- financial statements
- economic indicators
- portfolio holdings
