# Chapter 3 — Query Primer

Book: Learning SQL

Focus: Basic SELECT queries

## Learning Objectives

- Practice basic SELECT queries
- Use column aliases
- Use DISTINCT
- Filter rows with WHERE
- Sort results with ORDER BY
- Begin reading joins

## Query 1 — Select All Columns

````python
SELECT *
FROM language;
````

## Query 2 — Select Specific Columns

````python
SELECT language_id, name
FROM language;
````

## Query 3 — Column Alias

````python
SELECT
    language_id,
    upper(name) AS language_name
FROM language;
````

## Query 4 — Remove Duplicates

````python
SELECT DISTINCT actor_id
FROM film_actor
ORDER BY actor_id;
````

## Query 5 — Filter Rows

````python
SELECT title, rating, rental_duration
FROM film
WHERE rating = 'G'
  AND rental_duration >= 7;
````

## Query 6 — Sort Results

````python
SELECT title, rating, rental_duration
FROM film
WHERE rating = 'G'
ORDER BY rental_duration DESC, title;
````

## Query 7 — Basic Join Preview

````python
SELECT
    c.first_name,
    c.last_name,
    time(r.rental_date) AS rental_time
FROM customer AS c
INNER JOIN rental AS r
    ON c.customer_id = r.customer_id
WHERE date(r.rental_date) = '2005-06-14'
ORDER BY rental_time DESC;
````

## Observations

- `SELECT *` is useful for exploration, but specific columns are cleaner.
- Aliases make result sets easier to understand.
- `WHERE` narrows the dataset.
- `ORDER BY` controls presentation.
- Joins connect related tables through shared keys.

## Market Analysis Connection

The same query structure can be used for market data:

- select tickers and company names
